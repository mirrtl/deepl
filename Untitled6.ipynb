```python
from aiogram import Bot, Dispatcher
from aiogram.client.default import DefaultBotProperties
from aiogram.enums import ParseMode
from aiogram.fsm.storage.memory import MemoryStorage
from aiogram import Router, F
from aiogram.filters import CommandStart, Command, CommandObject
from aiogram.types import Message
from aiogram.types import KeyboardButton, ReplyKeyboardMarkup, KeyboardButtonPollType
from aiogram.utils.keyboard import ReplyKeyboardBuilder
from aiogram.types import InlineKeyboardMarkup, InlineKeyboardButton, WebAppInfo
from aiogram.utils.keyboard import InlineKeyboardBuilder
from aiogram.types import BotCommand, BotCommandScopeDefault
from aiogram.types import CallbackQuery
from aiogram.utils.chat_action import ChatActionSender
from aiogram.types import ReplyKeyboardRemove
from aiogram.fsm.context import FSMContext
from aiogram.fsm.state import State, StatesGroup
from typing import Any, Awaitable, Callable, Dict
from aiogram import BaseMiddleware
from aiogram.types import TelegramObject
from html import escape
import asyncio
import logging
import aiosqlite
from apscheduler.schedulers.asyncio import AsyncIOScheduler

class SomeMiddleware(BaseMiddleware):
    async def __call__(
        self,
        handler: Callable[[TelegramObject, Dict[str, Any]], Awaitable[Any]],
        event: TelegramObject,
        data: Dict[str, Any]
    ) -> Any:

        print(data['event_update'])

        result = await handler(event, data)

        return result

#–û–ø—Ä–µ–¥–µ–ª–∏–º —Å–æ—Å—Ç–æ—è–Ω–∏—è:
class Form(StatesGroup):
    name = State()
    age = State()

logging.basicConfig(force=True, level=logging.INFO, format='%(asctime)s - %(name)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

bot = Bot(token=token, default=DefaultBotProperties(parse_mode=ParseMode.HTML))
dp = Dispatcher(storage=MemoryStorage()) #Dispatcher - —ç—Ç–æ –æ—Å–Ω–æ–≤–Ω–æ–π –æ–±—ä–µ–∫—Ç, –æ—Ç–≤–µ—á–∞—é—â–∏–π –∑–∞ –æ–±—Ä–∞–±–æ—Ç–∫—É –≤—Ö–æ–¥—è—â–∏—Ö —Å–æ–æ–±—â–µ–Ω–∏–π –∏ –¥—Ä—É–≥–∏—Ö –æ–±–Ω–æ–≤–ª–µ–Ω–∏–π, –ø–æ—Å—Ç—É–ø–∞—é—â–∏—Ö –æ—Ç Telegram. –ò–º–µ–Ω–Ω–æ —á–µ—Ä–µ–∑ –¥–∏—Å–ø–µ—Ç—á–µ—Ä –ø—Ä–æ—Ö–æ–¥—è—Ç –≤—Å–µ —Å–æ–æ–±—â–µ–Ω–∏—è –∏ –∫–æ–º–∞–Ω–¥—ã, –æ—Ç–ø—Ä–∞–≤–ª—è–µ–º—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º–∏ –±–æ—Ç–∞.
#storage=MemoryStorage() —É–∫–∞–∑—ã–≤–∞–µ—Ç, —á—Ç–æ –¥–ª—è —Ö—Ä–∞–Ω–µ–Ω–∏—è —Å–æ—Å—Ç–æ—è–Ω–∏—è –∫–æ–Ω–µ—á–Ω—ã—Ö –∞–≤—Ç–æ–º–∞—Ç–æ–≤ (FSM)

start_router = Router() #–û–±—ä–µ–∫—Ç-–∫–æ–ø–∏—è Dispatcher –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Ö–µ–Ω–¥–ª–µ—Ä–∞

@dp.message(CommandStart())
async def cmd_start(message: Message, state: FSMContext):
    await message.answer(f'–ü—Ä–∏–≤–µ—Ç, {message.from_user.last_name} {message.from_user.first_name}! –í–∞—à id={message.from_user.id}')
    await state.set_state(Form.name)
    await state.update_data(name=message.text)
    async with aiosqlite.connect("users.db") as db:
      await db.execute("INSERT INTO users VALUES (?, ?, ?)", (message.from_user.id, message.from_user.last_name, message.from_user.first_name))
      await db.commit()

@start_router.message(F.text == "neww")
@start_router.message(Command("new"))
async def cmd_new(message: Message):
    inline_kb_list = [
        [InlineKeyboardButton(text="–ú–æ–π Telegram", url='https://t.me/varn_alex/')],
        [InlineKeyboardButton(text="–ú–æ–π Telegram", url='tg://resolve?domain=varn_alex')],
        [InlineKeyboardButton(text="–í–µ–± –ø—Ä–∏–ª–æ–∂–µ–Ω–∏–µ", web_app=WebAppInfo(url="https://tgbot.ru/app"))],
        [InlineKeyboardButton(text="–°–æ–∑–¥–∞—Ç—å", callback_data='new')],
        [InlineKeyboardButton(text="–£–¥–∞–ª–∏—Ç—å", callback_data='del')]
    ]
    keyboard = InlineKeyboardMarkup(inline_keyboard=inline_kb_list)
    await message.answer('–ú—ã –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–ª–∏ /new', reply_markup=keyboard)

@start_router.callback_query(F.data == 'new')
async def send_new(call: CallbackQuery):
    await call.answer('–°–æ–∑–¥–∞—é –Ω–æ–≤–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è', show_alert=True)
    await call.message.answer("–ù–æ–≤—ã–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å —Å–æ–∑–¥–∞–Ω!")

@start_router.message(Command('del'), Form.name)
async def cmd_del(message: Message):
    kb_list = [
    [KeyboardButton(text="üìñ –£–¥–∞–ª–∏—Ç—å –ø–æ–ª–Ω–æ—Å—Ç—å—é"), KeyboardButton(text="üë§ –£–¥–∞–ª–∏—Ç—å id")],
    [KeyboardButton(text="üìù –£–¥–∞–ª–∏—Ç—å –¥–∞—Ç—É"), KeyboardButton(text="üìö –£–¥–∞–ª–∏—Ç—å –§–ò–û")]
]
    keyboard = ReplyKeyboardMarkup(keyboard=kb_list, resize_keyboard=True, one_time_keyboard=True, input_field_placeholder="–í–æ—Å–ø–æ–ª—å–∑—É–π—Ç–µ—Å—å –º–µ–Ω—é:")
    await message.answer('–í—ã–±–µ—Ä–∏—Ç–µ –¥–µ–π—Å—Ç–≤–∏–µ', reply_markup=keyboard)
    #resize_keyboard=True: –∫–ª–∞–≤–∏–∞—Ç—É—Ä–∞ –±—É–¥–µ—Ç –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏ –∏–∑–º–µ–Ω—è—Ç—å —Ä–∞–∑–º–µ—Ä.
    #one_time_keyboard=True: –∫–ª–∞–≤–∏–∞—Ç—É—Ä–∞ —Å–∫—Ä—ã–≤–∞–µ—Ç—Å—è –ø–æ—Å–ª–µ –æ–¥–Ω–æ–≥–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è.
    #input_field_placeholder: –∑–∞–º–µ–Ω—è–µ—Ç —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—É—é –ø–æ–¥–ø–∏—Å—å ¬´–ù–∞–ø–∏—Å–∞—Ç—å —Å–æ–æ–±—â–µ–Ω–∏–µ...¬ª –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫—É—é.

@start_router.message(F.text)
async def cmd_mes(message: Message):
    await message.answer(message.html_text, reply_markup=ReplyKeyboardRemove())

async def send_msg(dp):
  async with aiosqlite.connect('users.db') as db:
    async with db.execute("SELECT * FROM users") as cursor:
      async for row in cursor:
        await bot.send_message(chat_id=row[0], text='–°–æ–æ–±—â–µ–Ω–∏–µ –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è')

async def start_bot():
    commands = [BotCommand(command='start', description='–°—Ç–∞—Ä—Ç'),
                BotCommand(command='new', description='–ù–æ–≤—ã–π'),
                BotCommand(command='del', description='–£–¥–∞–ª–∏—Ç—å')]
    await bot.set_my_commands(commands, BotCommandScopeDefault())

async def start_db():
    # –ü–æ–¥–∫–ª—é—á–µ–Ω–∏–µ –∫ –±–∞–∑–µ –¥–∞–Ω–Ω—ã—Ö (–µ—Å–ª–∏ —Ñ–∞–π–ª –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç, –æ–Ω –±—É–¥–µ—Ç —Å–æ–∑–¥–∞–Ω)
    async with aiosqlite.connect('users.db') as db:
        # –°–æ–∑–¥–∞–Ω–∏–µ —Ç–∞–±–ª–∏—Ü—ã, –µ—Å–ª–∏ –æ–Ω–∞ –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç
        await db.execute('''
            CREATE TABLE IF NOT EXISTS users (
                id INTEGER,
                lastname TEXT,
                firstname TEXT
            )
        ''')
        # –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –∏–∑–º–µ–Ω–µ–Ω–∏–π
        await db.commit()

async def main(): #–û—Å–Ω–æ–≤–Ω–∞—è –∞—Å–∏–Ω—Ö—Ä–æ–Ω–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è, –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç –∑–∞–ø—É—Å–∫–∞—Ç—å—Å—è –ø—Ä–∏ —Å—Ç–∞—Ä—Ç–µ –±–æ—Ç–∞.
    scheduler = AsyncIOScheduler(timezone='Europe/Moscow')
    scheduler.add_job(send_msg, 'interval', seconds=20, args=(dp,))
    scheduler.start()
    start_router.message.outer_middleware(SomeMiddleware())
    dp.include_router(start_router) #–î–æ–±–∞–≤–ª—è–µ–º —Ä–æ—É—Ç–µ—Ä start_router –≤ –¥–∏—Å–ø–µ—Ç—á–µ—Ä dp. –≠—Ç–æ –ø–æ–∑–≤–æ–ª—è–µ—Ç –¥–∏—Å–ø–µ—Ç—á–µ—Ä—É –∑–Ω–∞—Ç—å –æ –≤—Å–µ—Ö –æ–±—Ä–∞–±–æ—Ç—á–∏–∫–∞—Ö –∫–æ–º–∞–Ω–¥, –∫–æ—Ç–æ—Ä—ã–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –≤ start_router.
    dp.startup.register(start_bot)
    dp.startup.register(start_db)
    try:
      print("–ë–æ—Ç –∑–∞–ø—É—â–µ–Ω...")
      await bot.delete_webhook(drop_pending_updates=True)
      await dp.start_polling(bot, allowed_updates=dp.resolve_used_update_types()) #–∞–ø—É—Å–∫–∞–µ–º –±–æ—Ç–∞ –≤ —Ä–µ–∂–∏–º–µ –æ–ø—Ä–æ—Å–∞ (polling). –ë–æ—Ç –Ω–∞—á–∏–Ω–∞–µ—Ç –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ –∑–∞–ø—Ä–∞—à–∏–≤–∞—Ç—å –æ–±–Ω–æ–≤–ª–µ–Ω–∏—è —Å —Å–µ—Ä–≤–µ—Ä–∞ Telegram –∏ –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞—Ç—å –∏—Ö
    finally:
      await bot.session.close()
      print("–ë–æ—Ç –æ—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω")

asyncio.run(main())
```
